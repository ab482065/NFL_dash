### Objective:

Write function that obtains subset of `nfl_offensive_stats.csv` that contains the `game_id`, `player`, and `pass_yds`. 

Isolate segment of subset data where player is **Aaron Rodgers**. 

#### Future Objectives:

Replace `game_id` with **opponent information** (column currently undecided). 

```python
def get_player_subset_of_dataset(player: str = "Aaron Rodgers", 
                                 columns_to_isolate: list = ["game_id", "player", "pass_yds"], ):
    PLAYER_NAME_MATCH = (dataset["player"] == player)
    print(f"Player is `{player}`.")
    return dataset[columns_to_isolate][PLAYER_NAME_MATCH]
```

```python
subset = get_player_subset_of_dataset().drop(columns=["player"])

subset
```
